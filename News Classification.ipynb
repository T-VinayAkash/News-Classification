```python
dataf = data[['headline','category']]
```

```python
dataf
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
x = np.array(dataf['headline'])
y = np.array(dataf['category'])
```

```python
m = CountVectorizer()
x = m.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest =train_test_split(x,y,test_size = 0.20)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model = MultinomialNB()
model.fit(xtrain,ytrain)
```

```python
dataf['headline'].loc[10]
```

```python
dataf['category'].loc[10]
```

```python
input = 'messi wins the football cup'
sport = m.transform([input]).toarray()
```

```python
model.predict(sport)
```
